# *💰 Bitcoin Market Summary*

```python
# PACKAGES
import math
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size':18})
```

```python
# Plot market summary of bitcoin
btc_price = np.load("btc_price.npy")
plt.plot(btc_price, color='red', lw=1.5)
plt.ylabel('Bitcoin Price')
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['${:,.0f}'.format(x) for x in current_values])
plt.grid()
plt.show()
```

```python
# DENOISING THE PRICE TAGS

# Write your code here

n = len(btc_price)

D = np.zeros((n-1,n))

for i in range(n-1):
    D[i][i] = 1
    D[i][i+1] = -1
    

b = np.vstack((np.vstack(btc_price),np.zeros((n-1,1))))
b = np.array(b,dtype = 'float')

lmd = 20

A = np.vstack((np.eye(n),math.sqrt(lmd)*D))
A = np.array(A,dtype = 'float')

btc_price_denoised = np.linalg.lstsq(A, b)[0]

plt.plot(btc_price, 'b.', marker='.',lw=1.5)
plt.plot(btc_price_denoised, color='red', lw=1.5)
plt.ylabel('Denoised bitcoin Price, lambda='+str(lmd))
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['${:,.0f}'.format(x) for x in current_values])
plt.grid()
plt.show()
    

```

How does the parameter $\lambda$ change the plots?<br>
For which value of $\lambda$ the data looks more realistic (denoised)?

---
Answer:
1) As we observed, by increasing  $\lambda$ estimated data become smoother and more continus because the objective function wants to make consecutive points closer to each other. In this case, by increasing lamda we loss more details of original data.
On the other hand if  $\lambda$ is too small (below 15 in this problem), the estimated data will be too similar to the original noisy data because the objective function just wants to reduce the diffrence between original and estimated data and doesn't care about noisy data discontinuity.<br>
2) for about  $\lambda$=20 denoised data is acceptable, it is not to smooth and not to noisy.

```python
# PACKAGES
import matplotlib.pyplot as plt
import numpy as np
!pip install tqdm
from tqdm import tqdm 

plt.rcParams['figure.figsize'] = [16, 8]
plt.rcParams.update({'font.size':18})
```

```python
def hutchinson1(A, m):
    """
      Hutchinson's estimator.
      Args:
          A: target symmetric matrix, size n x n
          m: # standard normal random vectors 
      Returns: estimated trace
    """
    # Write your code here
    
    v = np.zeros((len(A),m))
    #v = np.random.normal(0,0.1,(len(A),m))
    for i in range(m):
        v[:,i] = np.random.normal(0,0.1,len(A))
    tr = (((v.T).dot(A)).dot(v)).trace()
    #tr = 0
    #for i in range(m):
     #   tr+=np.sum(v[:,i].T.dot(A).dot(v[:,i]))

    

    return tr/m

def hutchinson2(A, m):
    """
    Hutchinson's estimator.
    Args:
        A: target symmetric matrix, size n x n
        m: # random sign vectors 
    Returns: estimated trace
    """
    # Write your code here
    
    v = np.zeros((len(A),m))
    #v = np.random.binomial(1,0.5,(len(A),m))*2-1
    for i in range(m):
        v[:,i] = np.random.binomial(1,0.5,len(A))*2-1
    
    tr = (((v.T).dot(A)).dot(v)).trace()
    #tr = 0
    #for i in range(m):
       # tr+=np.sum(v[:,i].T.dot(A).dot(v[:,i]))

    return tr/m    
```

```python
def relative_error(true, est):
  return np.abs((est-true)/true)
```

```python
# Test and evalutate the estimators
# DO NOT CHANGE ANYTHING IN THIS CELL
num_test = 500
errors1 = []
errors2 = []

m = 10000

for i in tqdm(range(num_test), desc='Running tests'):
    n = np.random.randint(10, 15)
    A = np.random.randn(n,n)
    A = (A + A.transpose()) / 2
    tr_est1 = hutchinson1(A, m)
    tr_est2 = hutchinson2(A, m)
    tr_true = np.trace(A)
    errors1.append(relative_error(tr_true, tr_est1))
    errors2.append(relative_error(tr_true, tr_est2))

# Report the errors statistically
print(f"Random vectors: standard normal, m={m}:")
print(f"Relative error = {np.mean(errors1)} +- {np.std(errors1)}")

print(f"Random vectors: sign, m={m}:")
print(f"Relative error = {np.mean(errors2)} +- {np.std(errors2)}")
```
